```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('MRI_and_CDRinfo_Values_X_train.csv')
df2 = pd.read_csv('CDR_Values_y_train.csv')
```

```python
df["CDRGLOB"] = df2["CDRGLOB"]
means = df.groupby('CDRGLOB')['HIPPOVOL'].mean()
stds = df.groupby('CDRGLOB')['HIPPOVOL'].std()

print("Ortalamalar: \n",means)
print("\n Standart Sapma: \n",stds)

plt.figure(figsize=(18, 10))
for num,i in enumerate(list(df.groupby('CDRGLOB').groups.keys())):
    plt.subplot(2, 2, num+1)
    plt.subplots_adjust(hspace=0.2)
    plt.hist(df[df["CDRGLOB"]==i]['HIPPOVOL'])
    plt.axvline(means[i], color='black')
    plt.axvline(means[i]+stds[i], color='red')
    plt.axvline(means[i]-stds[i], color='red')
    plt.text(means[i]-(0.2),0.5,'Ortalama: '+str(round(means[i],2)),rotation=90)
    plt.text(means[i]+stds[i]-(0.2),0.5,r'$+1\sigma$ ('+str(round(means[i]+stds[i],2))+')',rotation=90)
    plt.text(means[i]-stds[i]-(0.2),0.5,r'$-1\sigma$ ('+str(round(means[i]-stds[i],2))+')',rotation=90)
    plt.title("CDRGLOB Group: "+str(i))
plt.show()
```
